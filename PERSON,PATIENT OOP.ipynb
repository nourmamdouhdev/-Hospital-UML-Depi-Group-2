```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def view_info(self):
        return "Name: " + self.name + " | Age: " + str(self.age)
```

```python
X=Person("mohamed", 20)
X.view_info()
```

```python
class Patient(Person):
    def __init__(self, name, age, medical_record):
        Person.__init__(self, name, age)
        self.medical_record = medical_record

    def view_info(self):
        return "Name: " + self.name + " | Age: " + str(self.age) + " | Record: " + self.medical_record

    def view_record(self):
        return "Medical Record: " + self.medical_record
```

```python
Y = Patient("Ahmed", 22, "MR1023")
Y.view_info()
```
